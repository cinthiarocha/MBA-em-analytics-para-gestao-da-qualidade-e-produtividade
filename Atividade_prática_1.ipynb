```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Define uma semente aleatória para garantir a reprodutibilidade dos resultados
np.random.seed(42)
# Criação de dois conjuntos de dados fictícios, cada um com uma distribuição Gaussiana
data1= np.random.normal(loc=-5, scale=1, size = 300)# 300 pontos com média -5 e desvio padrão 1
data2= np.random.normal(loc=5, scale=1, size = 700)# 700 pontos com média 5 e desvio padrão 1

# Combina os dois conjuntos de dados para formar um único conjunto
data = np.concatenate([data1, data2])

# Visualiza o conjunto de dados original com um gráfico
plt.hist(data, bins=30, density=True)
plt.title("Sucos Originais")
plt.show()

# Cria e treina um modelo de mistura Gaussiana com dois componentes (dois picos)
gmm = GaussianMixture(n_components=2)
# Ajusta o modelo de dados
gmm.fit(data.reshape(-1,1))

# Usando o treinamento para gerar 1000 novos pontos de dados
samples = gmm.sample(1000)[0]

plt.hist(samples, bins=30, density=True, alpha = .5, color = 'green', label='Suco Primavera')
plt.hist(data, bins=30, density=True, alpha = .5, color = 'red', label='Suco Original')
plt.title("Comparação: Suco Primavera vs. Suco Original")
plt.legend()
plt.show()
```
